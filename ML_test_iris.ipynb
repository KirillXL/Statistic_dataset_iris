```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile
import os

# Путь к ZIP-файлу на Google Диске
zip_path = '/content/drive/MyDrive/Colab Notebooks/iris.zip'
extract_path = '/content/drive/MyDrive/Colab Notebooks/iris'

# Создаем директорию для распаковки
os.makedirs(extract_path, exist_ok=True)

# Распаковка ZIP-файла
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Просмотр содержимого распакованной директории
extacted_files = os.listdir(extract_path)
print(extacted_files)

```

```python
# Путь к CSV-файлу внутри распакованной директории
csv_file_path = os.path.join(extract_path, 'Iris.csv')

# Загрузка данных
data = pd.read_csv(csv_file_path)

# Просмотр первых строк датасета
print(data.head())

# Общая информация о датасете
print(data.info())

# Статистическое описание данных
print(data.describe())
```

```python
# Построение гистограммы для одного из признаков (например, SepalLengthCm)
plt.figure(figsize=(8, 6))
sns.histplot(data['SepalLengthCm'], bins=20, kde=True, color='blue')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Построение диаграммы рассеяния для двух признаков (SepalLengthCm и SepalWidthCm)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=data, hue='Species', palette='Set1')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.legend(title='Species')
plt.show()
```

```python
# Построение парных диаграмм рассеяния для всех числовых признаков
sns.pairplot(data, hue='Species', palette='Set1')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```
